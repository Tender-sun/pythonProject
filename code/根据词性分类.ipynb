#### 定义一个文件读取函数，以空格分开读成列表。并对列表元素去重

```python
def rem(path):
    with open(path,encoding='utf-8')as f:
        line = f.read().split()
    all_word={}
    for i in line:
        if i in all_word:
            line.remove(i)
        else:
            all_word[i] = 1
    return line
```
